```python
lst1 = [4,61,2,4,2]
lst2 = [3,5,1,99,32]
```

```python
def max_lists(list1, list2):
    lst3 = []
    zipped_list = zip(lst1, lst2)
    for value in zipped_list:
        lst3.append(max(value))
    return lst3
```

```python
max_lists(lst1,lst2)
```

```python
mat = [[1, 2], [3, 4], [5, 6]]
```

```python
def get_diagonal(mat):
    diag = []
    for i in range(len(mat)):
        if i <= range(len(mat))[-1]:
            diag.append(mat[i][i])
    return diag
```

```python
get_diagonal(mat)
```
