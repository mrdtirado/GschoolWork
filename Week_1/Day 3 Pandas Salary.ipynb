Rename the columns as ['Name', 'Job', 'Department', 'Salary', 'Date']

```python
cols = ('Name','Job','Department','Salary','Date')
```

Check the data types in each column using df.info(). Which column has the wrong data type? Use apply on the column to replace it with the right type.

```python
df.info()
```

```python
df.columns = cols
```

```python
df.head()
```

```python
df['Salary'] = df['Salary'].apply(lambda x: x.replace('$', ''))
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
cats = (str, str, str, float)
```

```python
zipped = dict(zip(cols, cats))
```

```python
df = df.astype(zipped)
```
