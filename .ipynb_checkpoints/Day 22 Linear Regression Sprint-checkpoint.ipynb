### EDA: Continuous Predictors


```python
warnings.filterwarnings('ignore')

fig, ax = plt.subplots(figsize=(14,3))

plot_univariate_smooth(ax, 
                       df['Limit'].values.reshape(-1,1), 
                       df['Balance'], 
                       bootstrap=100)
ax.set_title("Univariate Effect of Credit Limit on Bank Balance")
ax.set_xlabel("limit")
ax.set_ylabel("Balance")


```

Accounts with no money behave differently than accounts with money in them.

```python
# Remove Accounts with less than 0 balance
newdf = df[df['Balance'] > 200]
```

```python
warnings.filterwarnings('ignore')

fig, ax = plt.subplots(figsize=(14,3))

plot_univariate_smooth(ax, 
                       newdf['Limit'].values.reshape(-1,1), 
                       np.log(newdf['Balance']), 
                       bootstrap=100)
ax.set_title("Univariate Effect of Credit Limit on Bank Balance")
ax.set_xlabel("limit")
ax.set_ylabel("Balance")
```
